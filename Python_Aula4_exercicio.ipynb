```
import matplotlib.pyplot as plt
# score antes e depois
plt.scatter(data=df,x='Mem_Score_Before',y='Mem_Score_After');
plt.title('Antes e Depois')
plt.xlabel('Antes')
plt.ylabel('Depois')
x=[25.0, 110]
plt.plot(x,x,color='red')# reta onde x = y
```

```
# analise do xanax dosagem vs score mem
plt.figure(figsize=(15,5))
plt.scatter(data=xanax, x='Mem_Score_Before',y='Mem_Score_After', c='Dosage')

plt.colorbar()# c é numerico  e por isso colorbar e não legend

plt.title('Antes e Depois')
plt.xlabel('Antes')
plt.ylabel('Depois')

x=[25.0, 110]
plt.plot(x,x,color='red')# reta onde x = y
plt.show()
```

```
# analise do xanax dosagem vs score mem : tam pontos por idade
plt.figure(figsize=(15,5))
plt.scatter(data=xanax, s='age', x='Mem_Score_Before',y='Mem_Score_After', c='Dosage')

plt.colorbar()# c é numerico  e por isso colorbar e não legend

plt.title('Antes e Depois')
plt.xlabel('Antes')
plt.ylabel('Depois')

x=[25.0, 110]
plt.plot(x,x,color='red')# reta onde x = y
plt.show()
```

```
# analise do xanax dosagem vs score mem : tam pontos por idade
plt.figure(figsize=(15,5))
fig = plt.scatter(data=xanax, s='age', x='Mem_Score_Before',y='Mem_Score_After', c='Dosage')

plt.colorbar()# c é numerico  e por isso colorbar e não legend
handles, labels = fig.legend_elements(prop='sizes')
plt.legend(handles, labels, loc='center right', bbox_to_anchor=(1.25, 0.5), title ='tamanho')

plt.title('Antes e Depois')
plt.xlabel('Antes')
plt.ylabel('Depois')

x=[25.0, 110]
plt.plot(x,x,color='red')# reta onde x = y
plt.show()
```

```
# Histograma idade
plt.hist(df.age, bins=20)
plt.title('Histograma idade')
plt.xlabel('idade')
plt.ylabel('quantidade');
```

```
# Histograma por droga
plt.hist(xanax.age, bins=20, alpha= 0.7, label='xanax')
plt.hist(halcion.age, bins=20, alpha= 0.7, label='halcion')
#plt.hist(placebo.age, bins=20, alpha= 0.7)
plt.legend()
plt.title('Histograma droga')
plt.xlabel('idade')
plt.ylabel('quantidade');
```

```
# Densidade: idade por droga
import seaborn as sns
sns.distplot(xanax.age, hist=False, label='xanax')
sns.distplot(halcion.age, hist=False, label='halcion')
sns.distplot(placebo.age, hist=False, label='placebo')

plt.title('Grafico densidade de idade por droga')
plt.xlabel('idade')
plt.ylabel('densidade')
```

```
# score de memoria por tipo de memoria feliz ou triste

happy=df[df.Happy_Sad_group == 'H']
sad=df[df.Happy_Sad_group == 'S']

plt.scatter(data=happy, x='Mem_Score_Before', y='Mem_Score_After', label='happy')
plt.scatter(data=sad, x='Mem_Score_Before', y='Mem_Score_After', label='sad')

plt.legend()
plt.title('Antes e Depois')
#plt.xlabel('Antes')
#plt.ylabel('Depois')

x=[30,100]
plt.plot(x,x, color='red');
```

```
# histograma da diferença de score de memoria feliz e triste
plt.hist(happy.Diff, alpha= 0.5, bins=20, label='happy')
plt.hist(sad.Diff, alpha= 0.5, bins=20, label='sad');

plt.legend()
plt.title('Histograma de Diff por grupo de memória')
#plt.xlabel('Diff')
#plt.ylabel('Quantidade')
```

```
sns.distplot(happy.Diff, hist=False, label='happy')
sns.distplot(sad.Diff, hist=False, label='sad')

plt.legend()
plt.title('Histograma de Diff por grupo de memória')
plt.xlabel('Diff')
plt.ylabel('densidade');
```
